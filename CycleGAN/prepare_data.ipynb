```python
import pandas as pd
import os
import shutil

# Baca metadata CSV
df = pd.read_csv('data/metadata.csv')

# Loop melalui setiap baris
for index, row in df.iterrows():
    domain = row['domain']
    split = row['split']
    path = row['image_path']
    
    # Tentukan animal berdasarkan domain
    if domain == 'A (Horse)':
        animal = 'horse'
    elif domain == 'B (Zebra)':
        animal = 'zebra'
    else:
        continue  # Skip jika tidak sesuai
    
    # Tentukan folder tujuan berdasarkan split
    if split == 'train':
        dest_folder = f'data/train/{animal}'
    elif split == 'test':
        dest_folder = f'data/val/{animal}'
    else:
        continue  # Skip jika tidak train atau test
    
    # Path sumber dan tujuan
    src = f'data/{path}'
    dest = f'{dest_folder}/{os.path.basename(path)}'
    
    # Pindahkan file
    if os.path.exists(src):
        shutil.move(src, dest)
        print(f'Moved {src} to {dest}')
    else:
        print(f'File {src} not found')
```

```python
import random

# Split sebagian gambar horse dari train ke val
horse_train_dir = 'data/train/horse'
horse_val_dir = 'data/val/horse'

files = os.listdir(horse_train_dir)
random.shuffle(files)

# Ambil 20% untuk validation
num_val = int(0.2 * len(files))
val_files = files[:num_val]

for f in val_files:
    src = os.path.join(horse_train_dir, f)
    dest = os.path.join(horse_val_dir, f)
    shutil.move(src, dest)
    print(f'Moved {f} to val/horse')

print(f'Total horse in train: {len(os.listdir(horse_train_dir))}')
print(f'Total horse in val: {len(os.listdir(horse_val_dir))}')
```
